## Problem 2

```python
import matplotlib.pyplot as plt
```

```python
combined_average = ece.groupby(['Track', 'Gender', 'Hometown'])['Average'].mean().unstack()
print("\nCombined Average Grade by Track, Gender, and Hometown:")
print(combined_average)

fig, axes = plt.subplots(3, 1, figsize=(10, 18))


# Average Grade by Track
ece.groupby('Track')['Average'].mean().plot(kind='bar', ax=axes[0], color='#AEC6CF', title='Average Grade by Track', ylabel='Average Grade')

# Average Grade by Gender
ece.groupby('Gender')['Average'].mean().plot(kind='bar', ax=axes[1], color='#FFDFDD', title='Average Grade by Gender', ylabel='Average Grade')

# Average Grade by Hometown
ece.groupby('Hometown')['Average'].mean().plot(kind='bar', ax=axes[2], color='#A020F0', title='Average Grade by Hometown', ylabel='Average Grade')

# Adjust layout
plt.tight_layout()


plt.show()
```
